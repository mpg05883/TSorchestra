Count the number of datasets in each data config.

```python
cfg_paths = list(Path("../conf/data").glob("*.yaml"))

for path in cfg_paths:
    with open(path, "r") as f:
        cfg = yaml.safe_load(f)

    print(f"{path.name}: {len(cfg)} datasets")
```
