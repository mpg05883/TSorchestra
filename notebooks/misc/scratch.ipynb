Import packages.

```python
import pandas as pd
import yaml

from pathlib import Path
```

Read dataset configs.

```python
cfg_path = Path("../../conf/data/all.yaml")

with open(cfg_path, "r") as f:
    cfg = yaml.safe_load(f)

print(type(cfg))
print(cfg)
```

Read metadata.

```python
metadata_path = Path("../../src/data/meta/metadata.csv")

df = pd.read_csv(metadata_path)

print(f"shape: {df.shape}")
df.head()
```

Create a yaml file for each domain.

```python
domains = df["domain"].unique().tolist()

for domain in domains:
    filtered_df = df[df["domain"] == domain]

    names, terms = filtered_df["name"].tolist(), filtered_df["term"].tolist()

    allowed_pairs = set(zip(names, terms))

    filtered_yaml = [
        entry for entry in cfg if (entry.get("name"), entry.get("term")) in allowed_pairs
    ]
    
    print(f"{domain}: {len(filtered_yaml)}")

    with open(f"{domain.lower().replace('/', '_')}.yaml", "w") as f:
        yaml.safe_dump(filtered_yaml, f, sort_keys=False)
```
